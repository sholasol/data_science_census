## Perfoming data cleaning in the Age column

```python
population_data[population_data['Age'].isnull()]
```

```python
#get the data type of Age column
# Some data are float, empty and string
# The data has to be converted to integer 
# replace the empty Age with the median age 
```

```python
print(population_data['Age'].dtypes)
```

```python
population_data['Age'].unique()
#convert the ages to numeric
#population_data['Age'] = pd.to_numeric(population_data['Age'], errors='coerce')
```

```python
#convert ages to numeric values

population_data['Age'] = pd.to_numeric(population_data['Age'], errors='coerce')
```

```python
# from the above ages has two different data types (float and int and empty '')

#replace the '' with nan
population_data['Age'] = population_data['Age'].replace(' ', np.nan)
```

```python
#Find the median of the age
age_median = population_data['Age'].median()
```

```python
age_median
```

```python
#fill the empty spaces with median age and convert it to integer data type

population_data['Age'] = population_data['Age'].fillna(age_median.round()).round().astype(int)
```

```python
# convert '' to nan, then we can fill the age in the specific column with the median age
```

```python
# Round the float number using Round then conver float to integers 
population_data['Age'] = population_data['Age'].astype(float).round().astype(int)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

# import warnings
# warnings.filterwarnings('ignore')

sns.set(rc={'figure.figsize':(25,30)})
```

### employment Categorization

```python
# given the fact that there are different occupation group
# It would be informative to group the people into four main categories
# Children, Students, Employed and Unemployed

def cat_occupation(occupation):
    if occupation == 'Student':
        return 'Student'
    elif occupation =='Unemployed':
        return 'Unemployed'
    elif occupation == 'Child':
        return 'Children'
    else:
        return 'Employed'
    
# Create new employment category column in the data frame
population_data['Occupation Category'] = population_data['Occupation'].apply(cat_occupation)

# count the number of records in each group

occ_grp = population_data.groupby('Occupation Category').size()

# visualize the result
#plt.rcParams.update({'font.size': 18})
plt.bar(occ_grp.index, occ_grp.values)

plt.xlabel('Occupation categories', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.title(' Occupation Distribution', fontsize=18)
plt.show()
```

```python
# Population age group
age_rng = pd.interval_range(start=0, end=100, freq=10)


#create age group in the dataframe
population_data['Age Group'] = pd.cut(population_data['Age'], bins=age_rng)


age_grp = population_data.groupby(['Age Group', 'Gender']).size().unstack()

# bar chart
fig, ax = plt.subplots(figsize=(8, 12))

ax.barh(age_grp.index.astype(str), -age_grp['Male'], align='center', color='blue')
ax.barh(age_grp.index.astype(str), age_grp['Female'], align='center', color='orange')

ax.set_xlabel('No. of People')
ax.set_ylabel('Age')
ax.set_title('Population Age Distribution Pyramid')

ax.invert_yaxis()
ax.set_yticks(age_grp.index.astype(str))
ax.set_yticklabels(age_rng.astype(str))

# plot legend
ax.legend(['Male', 'Female'])

plt.show()
```
